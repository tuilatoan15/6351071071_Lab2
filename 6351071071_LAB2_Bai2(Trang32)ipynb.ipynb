#1. Import thư viện và dữ liệu

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Ví dụ dữ liệu Precision và Recall
labels = [f'Bài {i}' for i in range(1, 11)]
precision = np.array([0.86, 0.72, 0.76, 0.69, 0.81, 0.77, 0.90, 0.68, 0.74, 0.80])
recall =    np.array([0.89, 0.64, 0.74, 0.71, 0.79, 0.82, 0.88, 0.70, 0.73, 0.78])
```

#2. Tính F-measure

```python
# Tính F-measure theo công thức trung bình điều hòa
f_measure = 2 * (precision * recall) / (precision + recall)

# Tạo bảng kết quả trực quan
df_results = pd.DataFrame({
    'Bài tập': labels,
    'Precision': precision,
    'Recall': recall,
    'F-measure': f_measure
})
df_results
```

#3: Biểu đồ cột so sánh Precision, Recall, F-measure

```python
x = np.arange(len(labels))
width = 0.25

plt.figure(figsize=(12,6))
plt.bar(x - width, precision, width, label='Precision', color='skyblue')
plt.bar(x, recall, width, label='Recall', color='orange')
plt.bar(x + width, f_measure, width, label='F-measure', color='green')

plt.xticks(x, labels)
plt.ylim(0, 1)
plt.ylabel('Giá trị')
plt.title('So sánh Precision, Recall và F-measure (10 nhãn)')
plt.legend()
plt.show()
```

#4: Heatmap trực quan

```python
heatmap_data = np.array([precision, recall, f_measure])
methods = ['Precision', 'Recall', 'F-measure']

plt.figure(figsize=(10,4))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="YlGnBu", xticklabels=labels, yticklabels=methods)
plt.title("Heatmap trực quan của Precision, Recall và F-measure (10 nhãn)")
plt.show()
```

#5. Nhận xét

- F-measure là chỉ số tổng hợp, được tính từ Precision và Recall, dùng để so sánh hiệu quả giữa các phương pháp một cách trực quan và khách quan. Chỉ số này đặc biệt hữu ích khi một phương pháp có độ chính xác cao nhưng độ phủ thấp, hoặc ngược lại, vì F-measure sẽ cân bằng hai yếu tố này, phản ánh đúng hiệu suất tổng thể.

- Qua phân tích F-measure của nhiều bài tập/nhãn, có thể đánh giá phương pháp nào hoạt động tốt hơn, đồng thời dễ dàng so sánh giữa các thuật toán khác nhau mà không bị lệ thuộc vào riêng Precision hay Recall. Phương pháp trực quan hóa bằng biểu đồ cột hoặc heatmap giúp minh chứng hiệu quả của các phương pháp một cách trực quan, khoa học.
