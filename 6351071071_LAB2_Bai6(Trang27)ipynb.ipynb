#Câu 4

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Đọc dữ liệu Mushroom
data = pd.read_csv("mushrooms.csv")

# Tách thuộc tính quyết định và thuộc tính đầu vào
features = data.drop('class', axis=1)
labels = data['class']

print("Nguyễn Hữu Toàn – 6351071071")
print("Shape của features:", features.shape)
print("Shape của labels  :", labels.shape)
```

#Câu 5

```python
features_onehot = pd.get_dummies(features)

print("Nguyễn Hữu Toàn – 6351071071")
print("Shape của features sau khi one-hot:", features_onehot.shape)
```

#Câu 6

```python
X_train, X_test, y_train, y_test = train_test_split(features_onehot, labels, test_size=0.3, random_state=42)

print("Nguyễn Hữu Toàn – 6351071071")
print("Kích thước X_train:", X_train.shape)
print("Kích thước X_test :", X_test.shape)
print("Kích thước y_train:", y_train.shape)
print("Kích thước y_test :", y_test.shape)
```

#Câu 7

```python
from sklearn import tree
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

clf_id3 = tree.DecisionTreeClassifier(criterion='entropy', random_state=0)
clf_id3.fit(X_train, y_train)
id3_pred = clf_id3.predict(X_test)
id3_acc = metrics.accuracy_score(y_test, id3_pred)

print("Nguyễn Hữu Toàn – 6351071071")
print("ID3 Accuracy:", id3_acc)
print("Report:\n", metrics.classification_report(y_test, id3_pred))

# Ma trận nhầm lẫn
id3_cm = metrics.confusion_matrix(y_test, id3_pred)
plt.figure(figsize=(8,8))
sns.heatmap(id3_cm, annot=True, fmt=".3f", cmap='Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('ID3 Decision Tree Accuracy: {0}'.format(id3_acc))
plt.show()

# Biểu diễn cây quyết định
fig, ax = plt.subplots(figsize=(20,12))
tree.plot_tree(clf_id3, filled=True, fontsize=8)
plt.show()
```

#Câu 8

```python
clf_cart = tree.DecisionTreeClassifier(criterion='gini', random_state=0)
clf_cart.fit(X_train, y_train)
cart_pred = clf_cart.predict(X_test)
cart_acc = metrics.accuracy_score(y_test, cart_pred)

print("Nguyễn Hữu Toàn – 6351071071")
print("CART Accuracy:", cart_acc)
print("Report:\n", metrics.classification_report(y_test, cart_pred))

# Ma trận nhầm lẫn
cart_cm = metrics.confusion_matrix(y_test, cart_pred)
plt.figure(figsize=(8,8))
sns.heatmap(cart_cm, annot=True, fmt=".3f", cmap='Blues')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('CART Decision Tree Accuracy: {0}'.format(cart_acc))
plt.show()

# Biểu diễn cây CART
fig, ax = plt.subplots(figsize=(20,12))
tree.plot_tree(clf_cart, filled=True, fontsize=8)
plt.show()
```

#Câu 9

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
bayes_pred = gnb.fit(X_train, y_train).predict(X_test)
bayes_acc = metrics.accuracy_score(y_test, bayes_pred)

print("Nguyễn Hữu Toàn – 6351071071")
print("Naive Bayes Accuracy:", bayes_acc)
print("Report:\n", metrics.classification_report(y_test, bayes_pred))

# Ma trận nhầm lẫn
bayes_cm = metrics.confusion_matrix(y_test, bayes_pred)
plt.figure(figsize=(8,8))
sns.heatmap(bayes_cm, annot=True, fmt=".3f", cmap='Greens')
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.title('Naive Bayes Accuracy: {0}'.format(bayes_acc))
plt.show()
```
